```python
from pylab import plot, imshow, show, xlabel, ylabel, xlim, ylim, hot
from numpy import loadtxt, array, zeros, sqrt, pi, sin, linspace
import numpy as np
import time
import matplotlib.pyplot as plt
from IPython import display as disp  
```

```python
cita=1
lamda=5
k=(2.*pi)/lamda
x1=50
y1=50
nx=500
dimx=100
dltx=dimx/nx

x=linspace(0,500,1)
y=linspace(0,500,1)
xlim(0.,500)
ylim(0.,500)

salida=open("ondas_est.dat","w")

for i in range(0,501) :
    for j in range(0,501):
        r1=sqrt((x1-i*dltx)**2+(y1-j*dltx)**2)
        C=cita*sin(k*r1)
        salida.write("%0.2f "% (cita*sin(k*r1)))
    salida.write("\n")
salida.close()
print(C)
data = loadtxt("ondas_est.dat", float)
imshow(data,origin="lower", extent=[0,100,0,100])
xlabel("Interferencia en X cm")
ylabel("interferencia en Y cm")
print(data.shape)
imshow(data)
show()
```

```python
cita=1
lamda=5
k=(2.*pi)/lamda
x1=50
y1=50
x2=25
y2=25
nx=500
dimx=100
dltx=dimx/nx

x=linspace(0,500,1)
y=linspace(0,500,1)
xlim(0.,500)
ylim(0.,500)

salida=open("ondas_est.dat","w")

for i in range(0,501) :
    for j in range(0,501):
        r1=sqrt((x1-i*dltx)**2+(y1-j*dltx)**2)
        r2=sqrt((x2-i*dltx)**2+(y2-j*dltx)**2)
        C1=cita*sin(k*r1)
        C2=cita*sin(k*r2)
        C=C1+C2
        salida.write("%0.2f "% C)
    salida.write("\n")
salida.close()
print(C)
data = loadtxt("ondas_est.dat", float)
imshow(data,origin="lower", extent=[0,100,0,100])
xlabel("Interferencia en X cm")
ylabel("interferencia en Y cm")
print(data.shape)
imshow(data)
show()
```

```python
import numpy as np
info=np.zeros((200,200))
np.savetxt("array",info,fmt="%s")
```

```python
salida2=open("ondas_estanque.dat","w")
s=600
d=100
x=linspace(0,d,s)
y=linspace(0,d,s)
X,Y=np.meshgrid(x,y)
c1=1 #amplitud
w1=2
b1=5 #longitud de onda
k1=(2*pi)/b1
x1=50 #psicion inicial x
y1=50 #posicion inicial y
t=25 #tiempo
info=np.zeros((s,s,t),dtype=float)
dimx=100
nx=500
dx=dimx/nx

for t in range(0,t):
    r1=sqrt((X-x1)**2 +(Y-y1)**2)
    C=c1*sin(k1*r1+w1*t)
    plt.gca().cla()
    info[:,:,t]=C
    imshow(C,origin="lower", extent=[0, d, 0, d])
    hot()
    disp.clear_output(wait=True)
    disp.display(plt.gcf())
```

```python
for r in range (t):
    salida2.write('#matriz %d \n' %(r))
    for i in range(s):
        for j in range(s):
            salida2.write('%0.4f ' %(info[i,j,r]))
        salida2.write('\n')
salida2.close()
```
